```python
import pandas as pd
```

```python
df = pd.read_csv('2016 Stack Overflow Survey Responses.csv')
```

```python
df.head(2)
```

```python
df = df[['tech_do']]
```

```python
df.head()
```

```python
df.value_counts()
```

```python
language = df['tech_do'].str.split(';',expand = True)
```

```python
graph= language.stack().value_counts().head(20)
```

```python
graph.plot(kind = 'bar',figsize = (12,6))
```
